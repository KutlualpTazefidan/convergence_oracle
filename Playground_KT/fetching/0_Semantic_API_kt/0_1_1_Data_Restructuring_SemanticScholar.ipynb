```python
import pandas as pd
import os
import copy
```

```python
# Retrieve the data from saved csv files for each year and combine it into a single df

# Create an empty DataFrame to store the merged data
merged_data = pd.DataFrame()
# Iterate through CSV files in the directory

# the start and end year will be corrected based on file names
start_year=None
end_year=None
for filename in os.listdir("./data/"):
    entry_path = os.path.join("./data/", filename)

    if os.path.isfile(entry_path):
        parts = filename.split("_")
        year = int(parts[0])
        # Update start_year and end_year
        if start_year is None or year < start_year:
            start_year = year
        if end_year is None or year > end_year:
            end_year = year
        if filename.endswith(".csv"):
            temp_df = pd.read_csv("./data/"+ filename)
            merged_data = pd.concat([merged_data, temp_df], ignore_index=True)

merged_data.to_csv('./data/'+ str(start_year) + "-" + str(end_year) + "_" + parts[1]+ "_" + str(len(merged_data))+'.csv', index=False)
```

```python
def flatten_dict(d, parent_key='', separator='_'):
    flattened = {}
    for key, value in d.items():
        # print("flattening: ",key)
        new_key = parent_key + separator + key if parent_key else key
        # print("sub key:?",isinstance(value, dict))
        if isinstance(value, dict):
            for sub_key, sub_value in value.items():
                print("sub_key: ",sub_key)
                sub_new_key = new_key + separator + sub_key
                flattened[sub_new_key] = sub_value
        else:
            flattened[new_key] = value
    return flattened
```

```python
flattened_data = []  # A list to store flattened dictionaries
original_data = pd.DataFrame(merged_data)  # Original DataFrame
num_records = len(merged_data)  # Number of records in the 'data' list

# Loop through each record in 'data' and flatten it
for i in range(num_records):
    flattened_data.append(flatten_dict(merged_data))

# Create a DataFrame from the flattened data
flattened_df = pd.DataFrame(merged_data)

# flattened_df.to_csv("./temp_data/all_features.csv",index=False)
# Display the first few rows of the flattened DataFrame
# display(flattened_df.head())
```
